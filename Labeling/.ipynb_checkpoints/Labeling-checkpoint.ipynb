```python
import pandas as pd
import yfinance as yf
```

```python
df = pd.read_csv('BTCUSDT-1h-2021-10.csv')
df = df.drop(columns=['quote_volume', 'taker_buy_volume', 'taker_buy_quote_volume', 'ignore'])
df.head()
```

```python
df['long'] = 0
# df['short'] = 0
```

```python
def label_long(tp = 2, sl = 1, time_limit = 24):
    for i in range(df.shape[0] - time_limit - 1):
        for j in range(1, time_limit + 1):
            if df.loc[i, 'close']*(1 + (tp/100)) <= df.loc[i + j, 'high']:
                df.loc[i, 'long'] = 1
            elif df.loc[i, 'close']*(1 - (sl/100)) >= df.loc[i + j, 'low']:
                df.loc[i, 'long'] = -1      
                
def label_short(tp = 2, sl = 1, time_limit = 5):
    for i in range(df.shape[0] - time_limit - 1):
        for j in range(1, time_limit + 1):
            if df.loc[i, 'close']*(1 + (sl/100)) <= df.loc[i + j, 'high']:
                df.loc[i, 'short'] = -1
            elif df.loc[i, 'close']*(1 - (tp/100)) >= df.loc[i + j, 'low']:
                df.loc[i, 'short'] = 1     
```
